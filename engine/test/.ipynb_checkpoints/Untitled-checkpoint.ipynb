```python
import sys
import cv2
import numpy as np
import random

from matplotlib import pyplot as plt
%matplotlib inline

def imshow(tit, image) :
    plt.title(tit)    
    if len(image.shape) == 3 :
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    else :
        plt.imshow(image, cmap="gray")
    plt.show()

```

```python
try:
    cap = cv2.VideoCapture(0)
    print('camera on')
except:
    print('fail')
    
lower_blue = (63, 79, 93)
upper_blue = (179, 255, 255)

lower_green = (0, 107, 197)
upper_green = (179, 255, 255)

lower_red = (-5, 100, 100)
upper_red = (5, 255, 255)

lower_orange = (6, 100, 100)
upper_orange = (13, 255, 255)

kernel = cv2.getStructuringElement(cv2.MORPH_CROSS, (3,3))

canny_min = 100
canny_max = 150

while(1):

    ret, frame = cap.read()
    
    blur = cv2.GaussianBlur(frame, (7,7), 7)
    hsv = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)
    
    # inRange
    mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
    mask_green = cv2.inRange(hsv, lower_green, upper_green)
    mask_red = cv2.inRange(hsv, lower_red, upper_red)
    
    
    # bitwise_and
    blue_object = cv2.bitwise_and(frame, frame, mask = mask_blue)
    green_object = cv2.bitwise_and(frame, frame, mask = mask_green)
    red_object = cv2.bitwise_and(frame, frame, mask = mask_red)
    
    
    # Canny
    blue_edge = cv2.Canny(blue_object, canny_min, canny_max)
    green_edge = cv2.Canny(green_object, canny_min, canny_max)
    red_edge = cv2.Canny(red_object, canny_min, canny_max)
    
    
    # morphology
    blue_edge_thick = cv2.morphologyEx(blue_edge, cv2.MORPH_CLOSE, kernel, iterations=3 )
    green_edge_thick = cv2.morphologyEx(green_edge, cv2.MORPH_CLOSE, kernel, iterations=3 )
    red_edge_thick = cv2.morphologyEx(red_edge, cv2.MORPH_CLOSE, kernel, iterations=3 )

    
    # Contour
    contours_b, _ = cv2.findContours(blue_edge_thick, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours_g, _ = cv2.findContours(green_edge_thick, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours_r, _ = cv2.findContours(red_edge_thick, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    for cnt in contours_b:
        area = cv2.contourArea(cnt) 
        cv2.drawContours(frame, [cnt], 0, (0,0,0), 3)
        if(area > 300): 
            x, y, w, h = cv2.boundingRect(cnt) 
            frame = cv2.rectangle(frame, (int(x*0.9), int(y*0.9)),
                                       (int((x + w)*1.1), int((y + h)*1.1)), 
                                       (255, 0, 0), 2)
            cv2.putText(frame, "Blue Colour", (x, y), 
                        cv2.FONT_HERSHEY_SIMPLEX, 
                        1.0, (255, 0, 0)) 

              
    for cnt in contours_g:
        cv2.drawContours(frame, [cnt], 0, (0,0,0), 3)
        
    for cnt in contours_r:
        cv2.drawContours(frame, [cnt], 0, (0,0,0), 3)
    
    

    cv2.imshow('src', frame)
#     cv2.imshow('blue', blue_object)
#     cv2.imshow('green', dst2)
#     cv2.imshow('red', dst3)
    
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
